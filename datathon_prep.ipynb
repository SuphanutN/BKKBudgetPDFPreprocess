```python
# Using readlines() 
file1 = open(file_path, 'r', encoding="utf8") 
Lines = file1.readlines() 
  
topic_list = []
price_list = []
flag = False
topic = ""
sub_topic_list = []
check_topic = ""

program_list = []
project_list = []
project = ""
project_flag = False

all_line = []
content_list = []
content = False

sub_topic = ""

sub_topic_a_flag = False
sub_topic_b_flag = False
sub_topic_c_flag = False
sub_topic_d_flag = False
sub_topic_e_flag = False
sub_topic_f_flag = False
sub_topic_g_flag = False
sub_topic_h_flag = False
sub_topic_i_flag = False
sub_topic_j_flag = False

# Strips the newline character 
for line in Lines: 
    #print(line)
    
    all_line.append(line)
    raw_line = line
    if line.find("(") == 0 and line[1] in ["1","2","3","4","5","6","7","8","9"]  :
        if line != "(ระยะที่ 2)                                                  90,550,000 บาท":
            line = line.replace('/n', '')
            line = line.replace('รวม ', '')
            line = line.replace(' า', 'ำ')
            line = line.replace('ซอื่', 'ซื่อ')
            line = line.strip()
            topic_list.append(line)
            flag = True
            topic = line
    elif flag:
        line = line.replace('/n', '')
        line = line.replace('รวม ', '')
        line = line.replace(' บาท', '')
        line = line.replace(',', '')
        line = line.strip()
        line = line.split(" ")
        price = line
        #print(line)
        #if "1" in line:
        if(len(line)==1):
            price_list.append(line[0])
        elif(len(line)>1):
            last = len(line)
            price_list.append(line[last-1])
        flag = False
        
    if check_topic != topic:
        if program_list != []:
            sub_topic_list.append({"topic": check_topic, "budget": price[0], "program_list": program_list, "content": content_list})
            program_list = []
            content_list = []
            content = False
            
    raw_line = raw_line.replace('แผนงานจัดการระบายน้ าและแก้ไขปัญหาน้ าท่วม', 'แผนงานจัดการระบายน้ำและแก้ไขปัญหาน้ำท่วม')
    raw_line = raw_line.replace('แผนงานพัฒนาระบบระบายน้ า', 'แผนงานพัฒนาระบบระบายน้ำ')
    raw_line = raw_line.replace('แผนงานจัดการคุณภาพน้ า', 'แผนงานจัดการคุณภาพน้ำ')
    
    if raw_line.find("ก.") == 0:
        sub_topic_a = raw_line.split("  ")
        sub_topic_a = sub_topic_a[1].strip()
        sub_topic_a_flag = True
        check_topic = topic
        sub_topic = sub_topic_a
    elif sub_topic_a_flag:
        sub_price_a = raw_line.replace('/n', '')
        sub_price_a = sub_price_a.replace(' บาท\n', '')
        sub_price_a = sub_price_a.replace(',', '')
        program_list.append({"program_name": sub_topic_a, "budget": sub_price_a})
        sub_topic_a = ""
        sub_price_a = ""
        sub_topic_a_flag = False
        content = True
        
    if raw_line.find("ข.") == 0:
        sub_topic_b = raw_line.split("  ")
        sub_topic_b = sub_topic_b[1].strip()
        sub_topic_b_flag = True
        sub_topic = sub_topic_b
    elif sub_topic_b_flag:
        sub_price_b = raw_line.replace('/n', '')
        sub_price_b = sub_price_b.replace(' บาท\n', '')
        sub_price_b = sub_price_b.replace(',', '')
        program_list.append({"program_name": sub_topic_b, "budget": sub_price_b})
        sub_topic_b = ""
        sub_price_b = ""
        sub_topic_b_flag = False
        content = True
        
    if raw_line.find("ค.") == 0:
        sub_topic_c = raw_line.split("  ")
        sub_topic_c = sub_topic_c[1].strip()
        sub_topic_c_flag = True
        sub_topic = sub_topic_c
    elif sub_topic_c_flag:
        sub_price_c = raw_line.replace('/n', '')
        sub_price_c = sub_price_c.replace(' บาท\n', '')
        sub_price_c = sub_price_c.replace(',', '')
        program_list.append({"program_name": sub_topic_c, "budget": sub_price_c})
        sub_topic_c = ""
        sub_price_c = ""
        sub_topic_c_flag = False
        content = True
        
    if raw_line.find("ง.") == 0:
        sub_topic_d = raw_line.split("  ")
        sub_topic_d = sub_topic_d[1].strip()
        sub_topic_d_flag = True
        sub_topic = sub_topic_d
    elif sub_topic_d_flag:
        sub_price_d = raw_line.replace('/n', '')
        sub_price_d = sub_price_d.replace(' บาท\n', '')
        sub_price_d = sub_price_d.replace(',', '')
        program_list.append({"program_name": sub_topic_d, "budget": sub_price_d})
        sub_topic_d = ""
        sub_price_d = ""
        sub_topic_d_flag = False
        content = True
        
    if raw_line.find("จ.") == 0:
        sub_topic_e = raw_line.split("  ")
        sub_topic_e = sub_topic_e[1].strip()
        sub_topic_e_flag = True
        sub_topic = sub_topic_e
    elif sub_topic_e_flag:
        sub_price_e = raw_line.replace('/n', '')
        sub_price_e = sub_price_e.replace(' บาท\n', '')
        sub_price_e = sub_price_e.replace(',', '')
        program_list.append({"program_name": sub_topic_e, "budget": sub_price_e})
        sub_topic_e = ""
        sub_price_e = ""
        sub_topic_e_flag = False
        content = True
        
    if raw_line.find("ฉ.") == 0:
        sub_topic_f = raw_line.split("  ")
        sub_topic_f = sub_topic_f[1].strip()
        sub_topic_f_flag = True
        sub_topic = sub_topic_f
    elif sub_topic_f_flag:
        sub_price_f = raw_line.replace('/n', '')
        sub_price_f = sub_price_f.replace(' บาท\n', '')
        sub_price_f = sub_price_f.replace(',', '')
        program_list.append({"program_name": sub_topic_f, "budget": sub_price_f})
        sub_topic_f = ""
        sub_price_f = ""
        sub_topic_f_flag = False
        content = True
        
    if raw_line.find("ช.") == 0:
        sub_topic_g = raw_line.split("  ")
        sub_topic_g = sub_topic_g[1].strip()
        sub_topic_g_flag = True
        sub_topic = sub_topic_g
    elif sub_topic_g_flag:
        sub_price_g = raw_line.replace('/n', '')
        sub_price_g = sub_price_g.replace(' บาท\n', '')
        sub_price_g = sub_price_g.replace(',', '')
        program_list.append({"program_name": sub_topic_g, "budget": sub_price_g})
        sub_topic_g = ""
        sub_price_g = ""
        sub_topic_g_flag = False
        content = True
        
    if raw_line.find("ซ.") == 0:
        sub_topic_h = raw_line.split("  ")
        sub_topic_h = sub_topic_h[1].strip()
        sub_topic_h_flag = True
        sub_topic = sub_topic_h
    elif sub_topic_h_flag:
        sub_price_h = raw_line.replace('/n', '')
        sub_price_h = sub_price_h.replace(' บาท\n', '')
        sub_price_h = sub_price_h.replace(',', '')
        program_list.append({"program_name": sub_topic_h, "budget": sub_price_h})
        sub_topic_h = ""
        sub_price_h = ""
        sub_topic_h_flag = False
        content = True
        
    if raw_line.find("ฌ.") == 0:
        sub_topic_i = raw_line.split(" ")
        sub_topic_i = sub_topic_i[1].strip()
        sub_topic_i_flag = True
        sub_topic = sub_topic_i
    elif sub_topic_i_flag:
        sub_price_i = raw_line.replace('/n', '')
        sub_price_i = sub_price_i.replace(' บาท\n', '')
        sub_price_i = sub_price_i.replace(',', '')
        program_list.append({"program_name": sub_topic_i, "budget": sub_price_i})
        sub_topic_i = ""
        sub_price_i = ""
        sub_topic_i_flag = False
        content = True
        
    if raw_line.find("ญ.") == 0:
        sub_topic_j = raw_line.split(" ")
        sub_topic_j = sub_topic_j[1].strip()
        sub_topic_j_flag = True
        sub_topic = sub_topic_j
    elif sub_topic_j_flag:
        sub_price_j = raw_line.replace('/n', '')
        sub_price_j = sub_price_j.replace(' บาท\n', '')
        sub_price_j = sub_price_j.replace(',', '')
        program_list.append({"program_name": sub_topic_j, "budget": sub_price_j})
        sub_topic_j = ""
        sub_price_j = ""
        sub_topic_j_flag = False
        content = True
        
    #if content == True:
    #    content_list.append(raw_line)
        
        
    ######
    
    if (raw_line.find("โครงการ") != -1 or raw_line.find("งาน") == 0 or raw_line.find("1.  งาน") == 0 
        or raw_line.find("2.  งาน") == 0 or raw_line.find("3.  งาน") == 0 
        or raw_line.find("4.  งาน") == 0 or raw_line.find("5.  งาน") == 0) and project_flag == False:
        project += raw_line
        project_flag = True
    elif project_flag == True :
        if raw_line.strip()[-3:] == "บาท":
            project += raw_line
            project_list.append([project,sub_topic,topic])
            project = ""
            project_flag = False
        else:
            project += raw_line
            
            
      
print(sub_topic_list)
#print(topic_list)

prep_topic_list = []
prep_price_list = []

for i in range(0,len(topic_list)):
    if (price_list[i]) is not "":
        try:
            print(int(price_list[i]))
            print(price_list[i].isdigit())
            prep_topic_list.append(topic_list[i])
            prep_price_list.append(int(price_list[i]))
        except:
            pass

print(prep_price_list)
#print(topic_list)
        
df = pd.DataFrame(list(zip(prep_topic_list, prep_price_list)), columns =['topic', 'price']) 

df = df.dropna(how='all')
```
